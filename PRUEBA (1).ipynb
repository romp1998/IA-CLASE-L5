```python
filtered_data = lista[(lista['AVG'] <= 270) & (lista['H'] <= 100000)]

colores=['green','red']
tamanios=[30,60]

f1 = filtered_data['AVG'].values
f2 = filtered_data['H'].values

asignar=[]
for index, row in filtered_data.iterrows():
    if(row['AVG']>0.270):
        asignar.append(colores[0])
    else:
        asignar.append(colores[1])   
plt.scatter(f1, f2, c=asignar, s=tamanios[0])
plt.show()
```
